```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import seaborn as sns
from wiswUtils import read_circle_patches
```

```python
INPUT_DATA_DIR = 'input_data/'
MATCHES_DATA_DIR = 'output_data/'
fnames = sorted([f for f in os.listdir(os.path.join(INPUT_DATA_DIR))  if f.endswith('csv') ])
patches_fnames = [os.path.join(INPUT_DATA_DIR, f) for f in fnames if 'patches' in f]
ori_fnames = [os.path.join(INPUT_DATA_DIR, f) for f  in fnames if 'ori' in f]
match_fnames_1 = sorted([os.path.join(MATCHES_DATA_DIR, f)  for f in os.listdir(os.path.join(MATCHES_DATA_DIR))  if f.endswith('_hardnet.csv') ])
match_fnames_2 = sorted([os.path.join(MATCHES_DATA_DIR, f)  for f in os.listdir(os.path.join(MATCHES_DATA_DIR))  if f.endswith('orinethardnet.csv') ])

NUM = 10
```

```python
matches = []
for fn in match_fnames_1[:NUM]:
    matches.append(np.loadtxt(fn))
matches = np.concatenate(matches, axis=0)
matches[matches<-4] = -4
```
